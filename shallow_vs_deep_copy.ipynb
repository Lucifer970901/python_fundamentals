## Shallow copy and Deep copy

```python
## =,copy(),deepcopy()
lst1 = [1,2,3,4]
lst2 = lst1
lst1,lst2
```

```python
lst2[1] = 1000
lst1,lst2
```

both lists are referencing same memory location  - so if one changes other changes too

```python
#copy
lst1 = [1,2,3,4]
lst2 = lst1.copy() #it stores it in different memory location
lst2[1] = 1000
lst1,lst2
```

copy() - changed the list to different memory location
both lists are in different memory location so changing one list does not effect the another

```python
#shallow copy with nested list
lst1 = [[1,2,3,4],[5,6,7,8]]
lst2 = lst1.copy()
```

```python
lst1
```

```python
lst2
```

```python
lst1[1][0] = 100
```

```python
lst1
```

```python
lst2
```

in case of nested list, at 2 memory locations there are multiple objects in 2 items of nested list.
it is refering to the same object inside the nested list
this is shallow copy
append()  - if you are making the changes inside the object present in the list - it will get applied to the list otherwise there will be no changes

```python
lst1.append([2,3,4,5])
```

```python
lst1
```

```python
lst2
```

```python
##deep copy
import copy
lst1 = [[1,2,3,4],[5,6,7,8]]
lst2 = copy.deepcopy(lst2)
lst2[1][0] =100
```

```python
lst2
```

if normal list(no nested list) shallow copy  = deep copy
