## Python Generators VS Iterators

```python
lst = [1,2,3,4]
for i in lst:
    print(i)
```

lst above is iterable

```python
#iterator
lst1= iter(lst)
```

```python
for i in lst1: #an iterator is iterable
    print(i)
```

```python
type(lst1)
```

```python
try:
    next(lst1) #dynamic memory cell
except StopIteration:
    print("iteration is empty")
```

for loop handles the **StopIteration** error of an iterator that you get while using next()

```python
## Genrator
def square(n):
    for i in range(n):
        return i ** 2
square(3)
```

```python
## Genrator
def square(n):
    for i in range(n):
        yield i ** 2
square(3) #created a generator object, this in turn creates an iterator
```

yield - a generator which converts it into iterator
it will not only return with respect to one keyword but for all the iterations and it will save the values

```python
a = square(3)
a
```

```python
next(a)
```

#### Difference between iterator and generator
1. to create iterator - we use iter() and to create generator we use function with yield keyword
2. Generator uses yield keyword - it saves the local variable value it can also return it
3. Generator helps us to write fast and compact code
4. Iterator is much more memory efficient

```python
## closures
def main_welcome(msg):
    def sub_welcome_class():
        print("welcome the community!")
        print(msg)
        print("please contribute to the community!")
    return sub_welcome_class()

main_welcome("hello everyone!")
```

```python
## closures and initial decorators
def main_welcome(func):
    def sub_welcome_class():
        print("welcome the community!")
        func("this is opensource")
        print("please contribute to the community!")
    return sub_welcome_class()

main_welcome(print)
```

```python
## closures and initial decorators
def main_welcome(func):
    def sub_welcome_class():
        print("welcome the community!")
        print(func([1,2,3,4,5,6,7,8,9,0]))
        print("please contribute to the community!")
    return sub_welcome_class()

main_welcome(len)
```

create the function and pass it as parameter

```python
##  decorators
def main_welcome(func):
    def sub_welcome_class():
        print("welcome the community!")
        func()
        print("please contribute to the community!")
    return sub_welcome_class()

def community():
    print("this is new community")
```

```python
main_welcome(community)
```

```python
@main_welcome
def community():
    print("this is new community")
```
