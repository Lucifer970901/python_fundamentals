## Numba

just in time compiler for python - used with array, loops and functions to make the execution faster

```python
!pip install numba
```

```python
from numba import jit
import numpy as np
import time

x = np.arange(1000).reshape(100,10)

@jit(nopython = True)
def go_fast(a): #function is compiled and runs in machine code
    trace = 0.0
    for i in range(a.shape[0]):
        trace += np.tanh(a[i,i])
    return a + trace

start = time.time()
go_fast(x)
end = time.time()
print("Elapsed (with compilation) = %s"% (end - start))

 #now the function is compiled, re-time it executing from cache
start = time.time()
go_fast(x)
end = time.time()    
print("Elapsed (after compilation) = %s"% (end - start))
```
