## SHA Cryptography

```python
import hashlib
```

```python
str1 = "Kyle peterson"
#process of hashing
#first step is to encode
#then apply hashing algorithm
hashedval = hashlib.sha256(str1.encode())
print(hashedval)
```

```python
#convert the value to hexadecimal to actually see the hash key
hashedval.hexdigest()
```

### SHA384

```python
hashedval = hashlib.sha384(str1.encode())
hashedval.hexdigest() 
```

this will have 96 characters i.e 4 bytes(for each characters)* 96 = 384

### SHA224

```python
hashedval = hashlib.sha224(str1.encode())
hashedval.hexdigest() 
```

### SHA512

```python
hashedval = hashlib.sha512(str1.encode())
hashedval.hexdigest()
```

```python
hashedval = hashlib.sha1(str1.encode())
hashedval.hexdigest()
```
