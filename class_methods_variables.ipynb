## Class Methods and Variables

```python
class Car:
    base_price = 100000
    def __init__(self,windows,doors,power):
        self.windows = windows
        self.doors = doors
        self.power = power
    def what_base_price(self):
        print(f"base price is {self.base_price}")
```

```python
car1 = Car(4,5,2000)
car1.base_price #access the variable using instance
```

```python
Car.base_price #access the variable using class
```

if I want to add 10% of the inflation rate to the base price - use class method

```python
class Car:
    base_price = 100000
    def __init__(self,windows,doors,power):
        self.windows = windows
        self.doors = doors
        self.power = power
    def what_base_price(self):
        print(f"base price is {self.base_price}")
    @classmethod
    def revise_base_price(cls,inflation):
        cls.base_price = cls.base_price + cls.base_price * inflation
```

```python
Car.revise_base_price(0.10) #good practice
```

```python
Car.base_price
```

```python
car1 = Car(4,5,2000)
car1.base_price
```

## Static methods

```python
import datetime
now = datetime.datetime.now()
```

```python
class Car:
    base_price = 100000
    def __init__(self,windows,doors,power):
        self.windows = windows
        self.doors = doors
        self.power = power
    def what_base_price(self):
        print(f"base price is {self.base_price}")
    @classmethod
    def revise_base_price(cls,inflation):
        cls.base_price = cls.base_price + cls.base_price * inflation
    @staticmethod
    def check_year():
        if now.year == 2026:
            return True
        else:
            return False 
```

```python
Car.check_year()
```

```python
car1 = Car(4,5,2000)
```

```python
#update the base price only if the year is different
if (car1.check_year()):
    pass
else:
    Car.revise_base_price()
```
