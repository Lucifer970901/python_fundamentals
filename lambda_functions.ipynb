## Lambda functions

also called as **anonymous functions** or a function with no name. works faster than a normal function

```python
def addition(a,b):
    return a + b

addition(4,5)
```

```python
addition = lambda a,b:a+b
addition(12,34)
```

```python
#check if a number is even or not
def evenornot(num):
    if num % 2 ==0:
        return True
    
evenornot(11)
```

```python
even1 = lambda a:a%2==0
even1(12)
```

```python
addition = lambda x,y,z : x+y+z
addition(2,3,4)
```

```python
#for more than one output

evenorodd = lambda num: f"{num} is even" if num%2==0 else f"{num} is odd" 
evenorodd(19)
```

```python
evenorodd(12)
```
