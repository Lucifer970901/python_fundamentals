## Eval function

Evalueates python expression which are written as strings


```python
eval("5*5")
```

```python
eval("5*5+7//2")
```

```python
eval(input("enter the expression"))
```

```python
def square_num(num):
    return num ** 2
square_num(2)
```

```python
eval("square_num(2)")
```

### How does Eval work

1. Parse the python expression
2. complie into byte code
3. Evaluate the python expression
4. it will return the result

```python
var = compile("5*5","string>","eval")
var
```

```python
eval(var)
```

```python
#global parameters
x =10
eval("x+50+x**2",{"x":x})
```

```python
x = 100
z=  100
eval("x+z+w",{"x":x,"z":z,"w":1000})
```

```python
x = 100
z = 100
eval("x+z")
```

```python
#local expression
eval("a+b+c",{},{"a":100,"b":200,"c":1000})
```

```python
a =100
b = 200
eval("a<b")
```
