## Data Classes
provides a concise way to define classes to store data, it automatically generates severals methods like **init, repr, eq** based on the class attributes that you define. simplifying the process of creating and working with data focused classes

```python
class Person:
    def __init__(self,name,age):
        self.name = name
        self.age = age

person1 = Person("Megha",26)
print(person1)
```

```python
from dataclasses import dataclass

@dataclass
class Person:
    name: str
    age : int
    profession : str


person1 = Person("Megha",26,"scientist")
print(person1)
```

The @dataclass decorator automatically generates the methods for you
1. **init()** - Initializes the object and assigns the provided values to the attributes
2. **attr()** - Provides a string representation of the object
3. **eq()** Implements equality comparison between 2 objects of the class based on their attributes

```python
@dataclass
class Person:
    name: str
    age : int
    profession : str = "AI"

person1 = Person("Megha",28)
print(person1)
```

```python
person1.profession = "Full stack developer"
```

```python
@dataclass
class Point:
    x:int
    y:int
point = Point(3,4)
print(point)
```

```python
 #immutable class
@dataclass(frozen = True)
class Point:
    x:int
    y:int
point = Point(3,4)
print(point)
```

```python
point.x = 34
```

```python
# inhertance
@dataclass
class Person:
    name:str
    age:int

@dataclass
class Employee(Person):
    employee_id:str
    dept:str

person1 = Person("Peter",23)
emp = Employee("peter",23,'123','AI')
print(emp)
```

```python
# nested dataclass

@dataclass
class Address:
    street:str
    city:str
    zipcode:str

@dataclass
class Person:
    name:str
    age:int
    address:Address
address = Address('123 main street','austin','345235')
person2 = Person("Peter",23,address)
```
