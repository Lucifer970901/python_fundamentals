if the class has no fixed attributes, any number of attributes can be created by the developer

so, here is the right approach

use **__init__()** to fix the attributes within a class

```python
class Car:
    def __init__(self,windows,doors,enginetype): #create 3 attributes 
        self.windows = windows
        self.doors = doors
        self.enginetype = enginetype
    def self_driving(self):
        return f" this is a {self.enginetype} car"
```

```python
car1 = Car()
```

```python
dir(car1) #gives the list of supported attributes
```

```python
car1 = Car(5,4,"Petrol")
```

```python
car2 = Car(3,4,"Diesel")
```

```python
print(car1.windows)
print(car2.doors)
print(car2.enginetype)
```

```python
car1.self_driving()
```

## OOPs - Inheritance

```python
#car blueprint
class Car:
    def __init__ (self, doors, windows, enginetype):
        self.doors = doors
        self.windows = windows
        self.enginetype = enginetype
    def drive(self):
        print("ther person drive the car")
```

```python
car=Car(4,5,"Petrol")
```

```python
dir(car)
```

```python
car.windows
```

```python
car.drive()
```

```python
#audi is child class, car is parent class from which it has to inherit objects and methods
class Audi(Car):
    def __init__(self,doors, windows, enginetype,enableai):
        super().__init__(doors, windows, enginetype) #calling from parent class
        self.enableai = enableai
    def selfdriving(self):
        print("Audi supports elf driving")
```

```python
audiq7 = Audi(5,5,"Petrol",True)
dir(audiq7)
```

```python
audiq7.drive()
```

```python
audiq7.selfdriving()
```

## OOPs - Magic methods in classes

```python
## all class variables are public
# car blueprint
class Car:
    def __init__ (self, doors, windows, enginetype):
        self.doors = doors
        self.windows = windows
        self.enginetype = enginetype
    def drive(self):
        print("ther person drive the car")
```

```python
c=  Car(4,5,"Diesel") #creating the object
```

```python
print(c)
```

```python
dir(c)
```

all the methods that we see above are called magic method in class

```python
class Car:
    def __init__ (self, doors, windows, enginetype):
        self.doors = doors
        self.windows = windows
        self.enginetype = enginetype
    def __str__(self): #overriding the existing  method
        return "the object has been initialized"
    def drive(self):
        print("ther person drive the car")
```

```python
c = Car(4,5,"Petrol")
print(c)
```

```python
c.__sizeof__() #shows in bytes
```

```python
class Car:
    def __new__(self, doors, windows, enginetype):
        print("the object is initialized")
    def __init__ (self, doors, windows, enginetype):
        self.doors = doors
        self.windows = windows
        self.enginetype = enginetype
    def __str__(self): #overriding the existing  method
        return "the object has been initialized"
    def drive(self):
        print("ther person drive the car")
```

```python
c = Car(4,5,"Petrol")
print(c)
```

the first print statement is from __new__() method

## Python OOPs - Public, protected and private

```python
#all class variables are public
class Car:
    def __init__(self,doors, windows,enginetype):
        self.windows = windows
        self.doors = doors
        self.enginetype = enginetype
```

```python
audi = (4,5,"Petrol")
audi
```

```python
dir(audi)
```

```python
#all class variables are protected
class Car:
    def __init__(self,doors, windows,enginetype):
        self._windows = windows
        self._doors = doors
        self._enginetype = enginetype
```

```python
audi = (4,5,"Diesel")
dir(audi)
```

```python
class Truck(Car):
    def __init__(self, windows, doors,enginetype, horsepower):
        super().__init__(windows, doors, enginetype)
        self.horsepower = horsepower
```

```python
truck = Truck(4,4,"Diesel",4000)
dir(truck)
```

as the defined variables are protected, only the child class truck can override the values of the variable

```python
truck._doors = 5
truck._doors
```

```python
# private variables
class Car():
    def __init__(self, windows, doors,enginetype):
        self.__windows = windows
        self.__doors = doors
        self.__enginetype = enginetype
```

```python
audi = Car(4,4,"Diesel")
audi
```
