## Exception Handling in python

```python
a = b
```

```python
try:
    #code block where exceptions can occur
    a = b
except:
    print("some problem may have occured")
```

```python
try:
    a = b
except Exception as ex:
    print(ex)
```

```python
try:
    a = b
except NameError as ex1:
    print("the user has not defined the variable")
except Exception as ex:
    print(ex)
```

here **NameError** is derived from the Exception class all the child classes have to be defined before the main class for exceptions

```python
try:
    a = 1
    b = 's'
    c = a + b
except NameError as ex1:
    print("the user has not defined the variable")
except Exception as ex:
    print(ex)
```

```python
#try else block
try:
    a = int(input("enter the number 1"))
    b = int(input("enter the number 2"))
    c = a / b
    d = a * b
    e = a + b
    

except NameError as ex1:
    print("the user has not defined the variable")
except TypeError as ex2:
    print("try to make the datatype similar")
except ZeroDivisionError as ex3:
    print("please give a number greater than 0")
except Exception as ex:
    print(ex)
else:
    print(c)
    print(d)
    print(e)
```

```python
## try else and finally
try:
    a = int(input("enter the number 1"))
    b = int(input("enter the number 2"))
    c = a / b
    d = a * b
    e = a + b
    

except NameError as ex1:
    print("the user has not defined the variable")
except TypeError as ex2:
    print("try to make the datatype similar")
except ZeroDivisionError as ex3:
    print("please give a number greater than 0")
except Exception as ex:
    print(ex)
else:
    print(c)
    print(d)
    print(e)
finally:
    print("execution is done")
```

### Custom Exception

```python
class Error(Exception):
    pass

class dobException(Error):
    pass
```

```python
year = int(input("enter the year of birth"))
age = 2026 - year
try:
    if age <= 30 and age > 20:
        print("age is valid, you can apply for the exam")
    else:
        raise dobException
except dobException:
    print("the year age is not valid, you cannot proceed further")
```
