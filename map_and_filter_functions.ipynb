## Map functions in python

The map() function is one of Python’s built-in power tools for functional programming. It allows you to take a function and apply it to every single item in an iterable (like a list or a tuple) without writing a manual for loop.

```python
def even_or_odd(num):
    if num % 2 == 0:
        return f"{num} is even"
    else:
        return f"{num} is odd"

even_or_odd(12)
```

```python
lst = [1,2,3,4,5,6,7,8,9,24,56,78 ]
evenorodd = lambda num: f"{num} is even" if  num% 2==0 else f"{num} is odd"
list(map(evenorodd,lst))
```

## Filter function in python

If the map() function is the assembly line that transforms every item, the filter() function is the security guard at the gate. It evaluates every item in an iterable against a specific condition and only lets the ones that return True pass through.

```python
even = lambda num: num % 2 ==0
```

```python
lst = [1,2,3,4,5,6,7,8,9,0]
```

```python
list(filter(even,lst)) #allows only those which pass the condition
```

```python
#i can use lambda function within filter
list(filter(lambda num:num%2==0,lst))
```

```python
list(map(lambda num:num%2==0,lst))
```
