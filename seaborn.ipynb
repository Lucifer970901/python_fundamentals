## Seaborn Tutorial

### Ditribution plots

* distplot
* joinplot
* pairplot

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

tip is dependent feature others are independent feature


### correlation with heatmap

```python
df.info()
```

```python
df.describe()
```

```python
df.corr(numeric_only = True) #helps with feature selections, works only on numeric type
```

positive values means linear correlation

negative value means negatively correlation (where one value increases other decrease)

```python
df.dtypes
```

```python
sns.heatmap(df.corr(numeric_only=True))
```

if the values of 2 features are towards 1 or 0.9  you can choose one of the  feature among them.

### Univariate analysis

```python
#join plot
sns.jointplot(x ='tip',y = 'total_bill',data = df,kind = 'hex' )
```

higher the value of histogram, majority of the feature's values are concetrated there.
as the total_bill increases, tip will also increase 
from this diagram, many people have given tip of $2 with total bill between $10 to $20 

```python
sns.jointplot(x = 'tip',y = 'total_bill',data = df,kind = 'reg')
```

based on histogram, this will draw Probability Density Function(PDF) and also regression line.

### pairplot

also known as scatter plot, in which one variable in the same data row is a matched with another variable's values, like this: pairplot are just elaborations on this, showing all the variables paired with all other variables

```python
sns.pairplot(data = df)
```

```python
df['sex'].value_counts()
```

```python
sns.pairplot(data = df, hue = 'sex')
```

```python
df['smoker'].value_counts()
```

```python
sns.pairplot(data = df,hue = 'smoker')
```

### Distplot

helps us to check the distribuion of the columns feature

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['total_bill'])
```

```python
sns.distplot(df['tip'],kde = False,bins=10)
```

now we can see the counts on y axis

### Categorical plot

* boxplot
* voilin plot
* countplot

#### countplot

gives the count in each categorical values

```python
##countplot
sns.countplot(x ='sex', data = df)
```

```python
sns.countplot(x = 'day',data = df)
```

```python
sns.countplot(y = 'sex',data =df)
```

```python
sns.barplot(x = 'total_bill',y = 'sex',data = df)
```

```python
sns.barplot(y = 'total_bill',x = 'smoker',data= df)
```

```python
sns.boxplot(x = 'smoker',y = 'total_bill',data = df)
```

```python
sns.boxplot(x = 'day',y = 'total_bill',data = df)
```

```python
sns.boxplot(x = 'sex',y = 'total_bill',data = df)
```

```python
sns.boxplot(x = 'day',y='total_bill',data = df,palette = 'rainbow')
```

```python
sns.boxplot(data = df,orient = 'v')
```

```python
sns.boxplot(x = 'total_bill',y = 'day',hue = 'sex',data =df)
```

```python
sns.violinplot(y = 'day',x = 'total_bill',data = df,palette = 'rainbow')
```
