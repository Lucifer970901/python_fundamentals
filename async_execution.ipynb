## Asynchronous Execution in python

```python
import time

from concurrent.futures import ThreadPoolExecutor
def return_number(a):
    return a
```

```python
start = time.time()
with ThreadPoolExecutor(max_workers = None) as executor:
    for result in executor.map(return_number,range(100)):
        print(f"count: {result}")

print(f"the total time is : {time.time() - start }")
```

let's add sleep so that we can see the execution happening parallely

```python
import time

from concurrent.futures import ThreadPoolExecutor
def return_number(a):
    time.sleep(1)
    return a
```

```python
return_number(12)
```

```python
start = time.time()
with ThreadPoolExecutor(max_workers = None) as executor:
    for result in executor.map(return_number,range(100)):
        print(f"count: {result}")

print(f"the total time is : {time.time() - start }")
```

when max_workers = None, python uses a dynamic formula to choose the number of workers it has to spawn dynamically

```python
start = time.time()
with ThreadPoolExecutor(max_workers = 5) as executor:
    for result in executor.map(return_number,range(100)):
        print(f"count: {result}")

print(f"the total time is : {time.time() - start }")
```
