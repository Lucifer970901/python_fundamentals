```python
num = 24

if num % 2 == 0:
    print(f"{num} is even")
else:
    print(f"{num} is odd")
```

```python
def even_odd(num):
    if num % 2 == 0:
        return f"{num} is even"
    else:
        return f"{num} is odd"
even_odd(32)
```

### Difference between return and print

* print - prints the statement, does not return anything
* return - returns the value after executing the function

```python
#this function just prints does not return anything
def hello_world():
    print("hello welcome")

hello_world()
```

```python
val = hello_world()
print(val) #its not returning anything
```

```python
def hello_world():
    return("hello welcome")

hello_world()
```

```python
val = hello_world()
print(val) #now its returning it as the value of a function
```

```python
def add_number(num1,num2):
    return num1 + num2
val = add_number(12,23)
val
```

### Positional and keyword argument

* positional argument - variable defined in the funciton without any value defined 
* keyword argument - variable defined as key value pairs

```python
def hello(name,age = 28):
    print("My name is {} and age is {}".format(name,age))
hello()
```

```python
hello("megha")
```

In the above example, "name" is **positional Argument** and "age" is **keyword argument**

```python
def hello(*args,**kwargs):
    print(args)
    print(kwargs)
hello("Megha","Shree",age = 28,dob = 1997)
    
```

```python
lst = ['Megha','Shree']
dict_args = {'age': 28, 'dob': 1997}
hello(lst,dict_args)
```

```python
hello(*lst,**dict_args)
```

```python
#multiple values returned from a function
lst = [1,2,3,4,5,6,7]
def evenoddsum(lst):
    even_sum = 0
    odd_sum = 0
    for i in lst:
        if i % 2 ==0:
            even_sum = even_sum + i
        else:
            odd_sum = odd_sum + i
    return even_sum,odd_sum

evenoddsum(lst)
```

### is vs ==

```python
lst1 = [1,2,3,4]
lst2 = [1,2,3,4]
lst1 == lst2 #returns true even if its in different memory location as long as the item is same
```

```python
lst1 is lst2
```

```python
lst1 = [1,2,3,4]
lst2 = lst1  #compared the items in same memory location
lst1 is lst2
```
