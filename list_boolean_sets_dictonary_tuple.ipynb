```python
#string functions
string = "Megha1 shree"
print(string.isalnum()) #check if all characters are numbers
print(string.isalpha()) #check if all characters are alphabetic
print(string.isdigit()) # test if string contains digits
print(string.istitle()) #check for title words
print(string.isupper())
print(string.islower())
print(string.isspace()) #if it contains spaces
print(string.endswith('e'))
print(string.startswith('M'))
```

### sets

```python
set1 = {1,2,3,4,5,6,5,5,6}
print(set1)
print(type(set1))
```

```python
#indexing in set - not allowed as it is unordered
set1(1)
```

```python
set1.add("hulk")
set1
```

```python
set1 = {'avengers','ironman','hitman'}
set2 = {'avengers','ironman','hitman','hulk'}
```

```python
set2.difference(set1)
```

```python
set2.difference_update(set1)
```

```python
print(set2)
```

```python
set1 = {'avengers','ironman','hitman'}
set2 = {'avengers','ironman','hitman','hulk'}
set2.intersection(set1)
```

```python
print(set1,set2)
```

```python
set2.intersection_update(set1)
```

```python
print(set1,set2)
```

### Dictionaries

```python
dic = {}
type(dic)
```

```python
my_dict = {'car1':"Audi","car2":"BMW","Car3":"Mercidies Benz"}
print(type(my_dict))
```

```python
my_dict['car1']
```

```python
for x in my_dict:
    print(x)
```

```python
for x in my_dict.values():
    print(x)
```

```python
for x in my_dict.items(): #iterate through pairs (keys,values)
    print(x)
```

```python
my_dict['car4'] = 'Ford'
print(my_dict)
```

```python
my_dict['car1'] = 'Maruti' #replaces the value in that key
my_dict
```

### Nested dictionary

```python
car1_model = {'Mercedes':1960}
car2_model = {'Audi':1970}
car3_model = {'Ambassador':1980}
car_type = {'car1':car1_model,'car2':car2_model,'car3':car3_model}
```

```python
print(car_type)
```

```python
print(car_type['car1'])
```

```python
print(car_type['car1']['Mercedes'])
```
