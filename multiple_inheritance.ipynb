```python
class A():
    def method1(self):
        print("A class method is called")
```

```python
class B(A):
    def method1(self):
        print("B class method is called")
    def method2(self):
        print("c class method2 is called")    
```

```python
class C(A):
    def method1(self):
        print("c class method is called")
    
```

```python
class D(B,C):
    def method1(self):
        print("D class method is called")
```

```python
d = D()
d.method1()
d.method2()
```

```python
B.method1(d)
```

```python
A.method1(d)
```

```python
class D(B,C):
    def method1(self):
        print("D class method is called")
        C.method1(self)
        B.method1(self)
        A.method1(self)
```

```python
d = D()
d.method1()
```
