## List Comprehension

provides a consise way to create the list. consists of brackets containing an expression followed by for clause, then zero or more for or if clauses. the expressions can be anything, meaning you can put all kind of objects in the lists.

```python
lst1 = []
def lst_square(lst):
    for i in lst:
        lst1.append(i*i)
    return lst1

lst_square([1,2,3,4,5,6,7])
```

```python
lst = [1,2,3,4,5,6,7]
```

```python
[i*i for i in lst]
```

```python
lst1 = [i*i for i in lst]
lst1
```

```python
#return the square of number only if the number is even
even = [i*i for i in lst if i % 2 ==0]
even
```

```python
odd = [i*i for i in lst if i % 2 !=0]
odd
```
