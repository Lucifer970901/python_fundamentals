## Zip Function parallel iteration


```python
lst1 = ["krish","sam","John"]
lst2 = ["a"," b","c"]
```

```python
for i in lst1:
    print(i)
```

```python
op = zip(lst1,lst2)
op
```

```python
list(op)
```

```python
#zip is an iterator
next(op)
```

```python
op = zip(lst1,lst2)
for i,j in op:
    print(i,j)
```

```python
lst1 = ["krish","sam","John",4]
lst2 = ["a"," b","c"]
lst3 = [1,2,3]

output = zip(lst1,lst2,lst3)
for i,j,k in output:
    print(i,j,k)
```

```python
list(zip())
```

```python
dict1 = {"name":"megha","last_name":"nayak","age":28}
dict2 = {"name":"john","last_name":"peter","age":30}
dictionary = zip(dict1.items(),dict2.items())
```

```python
for (i,j),(i2,j2) in dictionary:
    print(i,j)
    print(i2,j2)
```
