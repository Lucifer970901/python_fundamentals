```python
from PyPDF2 import PdfReader, PdfWriter
```

```python
#open the current pdf
file_pdf = PdfReader('data/attention-is-all-you-need-Paper.pdf')
#object for pdf writer
out_pdf = PdfWriter()
```

```python
file_pdf
```

```python
len(file_pdf.pages)
```

```python
for i in range (len(file_pdf.pages)):
    page_details = file_pdf.pages[i]
    #add to the output page
    out_pdf.add_page(page_details)
```

```python
password = "Megha@123"
out_pdf.encrypt(password)
```

```python
with open("data/encrypted_paper.pdf","wb") as filename:
    out_pdf.write(filename)
```
