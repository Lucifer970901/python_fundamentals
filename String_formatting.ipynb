```python
#.format method - legacy method to store and pass the value to the string
def greeting(name):
    return "Hello {}. welcome to the community".format(name)
```

```python
# f strings
def welcome_email(firstname,age):
    return f"welcome {firstname} your age is {age} "
```

```python
welcome_email("megha",28)
```

```python
#another way with format
def welcome_email(name,age):
    return "welcome {name}.  your age is {age}".format(age = age,name = name)

welcome_email("megha",28)
```

```python
def welcome_email(name,age):
    return "welcome {name}.  your age is {age}".format(age = name,name = age)

welcome_email("megha",28)
```

```python
def welcome_email(name,age):
    return "welcome {name1}.  your age is {age1}".format(age1 = age,name1 = name)

welcome_email("megha",28)
```
