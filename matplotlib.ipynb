### subplots

```python
plt.subplot(2,2,1) #nrows,ncols,position
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*-')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.subplot(2,2,4)
plt.plot(x,y,'m*--')
```

```python
x = np.arange(1,11)
y = 3 * x + 5
plt.title("matplotlib demo")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.plot(x,y)
```

```python
#let's create sine wave form
np.pi
```

```python
x = np.arange(0,4*np.pi,0.1)
y = np.sin(x)
plt.title("sine wave form")

plt.plot(x,y)
```

```python
#subplots with sine and cos waves
x = np.arange(0,5*np.pi,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(2,1,1)
plt.plot(x,y_sin,'g--')
plt.title("sine ")

plt.subplot(2,1,2)
plt.plot(x,y_cos,'r--')
plt.title("cosine ")
```

```python
#bar plot

x = [2,8,10]
y = [11,16,9]

x2 = [3,9,11]
y2 = [6,15,7]
plt.bar(x,y)
plt.bar(x2,y2,color= 'g')
plt.title('bar graph')
plt.xlabel('x axis')
plt.ylabel('y axis')
```

```python
a = np.random.randint(0,100,20) #create 20 numbers from 0 to 100 range
plt.hist(a)
plt.title("histogram")
```

### Box plot in matplotlib

```python
data = [np.random.normal(0,std,100) for std in range(1,4)]
plt.boxplot(data,vert = True,patch_artist = True)
```
